```python
import pandas as pd
import numpy as np
```

```python
df_1980 = pd.read_csv("/content/angev98_1980_raw.csv")
```

```python
df_1990 = pd.read_csv("/content/angev98_1990_raw.csv")
```

```python
import re
def clean_keep_code(x):
    if pd.isna(x):
        return pd.NA
    s = str(x).strip()
    s = re.sub(r"""^b['"]?(.*?)['"]?$""", r"\1", s)
    s = re.sub(r"[^0-9]", "", s)
    return s if s != "" else pd.NA

df_clean_1980 = df_1980.apply(lambda col: col.map(clean_keep_code))

df_clean_1990 = df_1990.apply(lambda col: col.map(clean_keep_code))


print(df_clean_1980.head())
print(df_clean_1990.head())
```

```python
df_cleaned_1990 = df_clean_1990.dropna()
df_cleaned_1980 = df_clean_1980.dropna()
```
