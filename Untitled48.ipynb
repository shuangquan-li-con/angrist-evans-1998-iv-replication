```python
import pandas as pd

df_1980 = pd.read_sas("AngEv98/AngEv98/m_d_806.sas7bdat")
df_1990 = pd.read_sas("AngEv98/AngEv98/m_d_903.sas7bdat")

```

```python
print(df_1980.shape)
print(df_1990.shape)
```

```python

for col in df_1980.columns:
    if df_1980[col].dtype == object:
        df_1980[col] = df_1980[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

for col in df_1990.columns:
    if df_1990[col].dtype == object:
        df_1990[col] = df_1990[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```

```python
print(df_1990.head())
print(df_1980.head())
```

```python
df_1980.columns = df_1980.columns.str.lower()
df_1990.columns = df_1990.columns.str.lower()

print(df_1980.columns[:10])
```
